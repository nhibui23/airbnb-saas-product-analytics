## 3. Clean data
Check for missing values and count total missing values from each column in descending order

## Comment
* license → 102,597 nulls out of 102,599. Basically empty => redundant
* house_rules → half missing => not really needed 
* country / country code will be the same since it's all NYC => not needed column
* 102599 entries and 26 columns
* price and service fees should be numbers instead of strings

```python
#1. Drop unneeded column
df.drop(columns=['license', 'house_rules', 'country', 'country code'], inplace = True)
```

```python
#2. Change number of reviews and last review to 0
df['reviews per month'] = df['reviews per month'].fillna(0)
df['number of reviews'] = df['number of reviews'].fillna(0)
```

```python
#3. Drop rows with small missing value (<250)
df.dropna(subset = ['lat', 'long', 'neighbourhood', 'neighbourhood group', 'cancellation_policy', 'instant_bookable', 'number of reviews', 'Construction year', 'price'])
```

```python
#4. Change data type of price, service fee and last review
df['price']= df['price'].str.replace(['$,'], "", regex = True).astype(float)
```
